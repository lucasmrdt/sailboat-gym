```python
import sys
sys.path.append('..')

import matplotlib.pyplot as plt
import numpy as np
from sailboat_gym import load_best_sail_dict, get_best_sail
```

```python
best_sail_dict = load_best_sail_dict('SailboatLSAEnv-v0')
```

```python
thetas = np.array(list(best_sail_dict.keys()))
sail = np.array(list(best_sail_dict.values()))

thetas = np.rad2deg(thetas)
sail = np.rad2deg(sail)

plt.subplots(figsize=(3, 3), dpi=300)

plt.plot(thetas, sail)
plt.xlabel('Boat angle (deg)')
plt.ylabel('Sail angle (deg)')
plt.title('Best sail angle');
```

```python
thetas = np.deg2rad(np.arange(0, 360, 1))
sails = np.array([get_best_sail('SailboatLSAEnv-v0', theta, 2) for theta in thetas])

thetas = np.rad2deg(thetas)
sails = np.rad2deg(sails)

plt.subplots(figsize=(3, 3), dpi=300)

plt.scatter(thetas, sails, s=1, c='k')
plt.xlabel('Boat angle (deg)')
plt.ylabel('Sail angle (deg)')
plt.title('Best sail angle');
```
